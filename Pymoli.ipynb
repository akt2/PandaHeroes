```python
spdf=pdf.groupby('SN')
tspc=spdf['Price'].count()
# ael = pdf.loc[pdf['SN'] == "Aelastirin39"]
tstp=spdf['Price'].sum()
sortt=tstp.sort_values(ascending=False)
top=sortt.head(5)
```

```python
tdf=pd.DataFrame(top)
tdf=tdf.reset_index()
```

```python
spend=pdf[pdf['SN'].isin(tdf['SN'])].groupby('SN').agg({'Purchase ID': ['count'], 'Price': ['mean','sum']})
spend={'Purchase Count': spend['Purchase ID']['count'],'Average Purchase Price':spend['Price']['mean'],'Total Purchase Value':spend['Price']['sum']}
spend=pd.DataFrame(spend)
spend.style.format({'Average Purchase Price':"${:.2f}",'Total Purchase Value':'${:.2f}'})
```

## Most Popular Items

```python
pop=pdf.groupby(['Item ID','Item Name']).agg({'Purchase ID': ['count'], 'Price': ['mean','sum']})
pop=pd.DataFrame({'Purchase Count': pop['Purchase ID']['count'],'Average Purchase Price':pop['Price']['mean'],'Total Purchase Value':pop['Price']['sum']})
pop.style.format({'Average Purchase Price':"${:.2f}",'Total Purchase Value':'${:.2f}'})
```
